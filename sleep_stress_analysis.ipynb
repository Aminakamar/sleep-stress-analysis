```python
import pandas as pd
import matplotlib.pyplot as plt

# load data
df = pd.read_csv(list(uploaded.keys())[0])

df.head()
```

```python
df = df.dropna()
```

```python
plt.scatter(df['Screen Time'], df['Sleep Duration'])
plt.xlabel("Screen Time")
plt.ylabel("Sleep Duration")
plt.title("Screen Time vs Sleep")
plt.show()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Sleep Duration'], df['Stress Level'])
plt.xlabel("Sleep Duration")
plt.ylabel("Stress Level")
plt.title("Sleep vs Stress")
plt.show()
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['daily_screen_time_hours'], df['sleep_duration_hours'])
plt.xlabel("Screen Time (hours)")
plt.ylabel("Sleep Duration (hours)")
plt.title("Screen Time vs Sleep")
plt.show()
```

```python
plt.scatter(df['daily_screen_time_hours'], df['stress_level'])
plt.xlabel("Screen Time (hours)")
plt.ylabel("Stress Level")
plt.title("Screen Time vs Stress")
plt.show()
```

```python
plt.scatter(df['sleep_duration_hours'], df['stress_level'])
plt.xlabel("Sleep Duration")
plt.ylabel("Stress Level")
plt.title("Sleep vs Stress")
plt.show()
```

```python
print("Average Screen Time:", df['daily_screen_time_hours'].mean())
print("Average Sleep:", df['sleep_duration_hours'].mean())
print("Average Stress:", df['stress_level'].mean())
```

```python
corr = df[['daily_screen_time_hours',
           'sleep_duration_hours',
           'stress_level']].corr()

print(corr)
```

```python
import matplotlib.pyplot as plt

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

Result:

The average screen time is moderately high (~5.5 hours/day).
Average sleep duration is slightly low (~6.5 hours).
Stress levels are relatively high (~7).

Conclusion:

Increased screen time is associated with reduced sleep duration.
Reduced sleep leads to higher stress levels.
Therefore, managing screen time can improve sleep and reduce stress.
